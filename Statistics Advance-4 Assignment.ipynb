```python
Q1: What is the difference between a t-test and a z-test? Provide an example scenario where you would use each type of test.

t-test: Used when the sample size is small (typically 
ùëõ<30) or when the population standard deviation is unknown. It accounts for the additional variability that arises from estimating the population standard deviation from the sample.
Example: Testing whether the mean weight of a group of 15 participants differs from a known population mean.
z-test: Used when the sample size is large (typically 
ùëõ‚â•30) and the population standard deviation is known. It assumes that the sample means are normally distributed.
Example: Testing whether the average score of 100 students is significantly different from a national average score, assuming the population standard deviation is known.
```

```python
Q2: Differentiate between one-tailed and two-tailed tests.

One-tailed test: Tests for a deviation in a specific direction (either greater than or less than a specified value).
Example: Testing if a new drug improves recovery time (you only care if it‚Äôs shorter, not longer).
Two-tailed test: Tests for a deviation in both directions (either greater or less than a specified value).
Example: Testing if a new teaching method leads to a significant difference in exam scores compared to the standard method, regardless of whether the scores go up or down.
```

```python
Q3: Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

Type 1 error: Rejecting the null hypothesis when it is actually true (false positive).
Example: A medical test indicating that a patient has a disease when they do not (false positive diagnosis).
Type 2 error: Failing to reject the null hypothesis when it is actually false (false negative).
Example: A medical test failing to detect a disease in a patient who actually has it (false negative).
```
